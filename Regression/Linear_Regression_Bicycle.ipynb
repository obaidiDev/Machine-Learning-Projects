```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
scaler = StandardScaler()
normalizer = Normalizer()
```

```python
data_df = pd.read_csv('bikers_data.csv')
```

```python
data_df.head()
```

```python
data_y = data_df['Number of bikers'] # target
data_x = data_df.drop(['Number of bikers','Date'], axis=1) # input features
```

```python
data_x.head()
```

```python
data_y
```

```python
data_df.drop("Date",axis=1).corr() # I will drop the date since it is not that important
# most of the data are highly correlated excpect friday but dropping it does not make sense,
# since we are considring other days
```

```python
X = data_x
X_train, X_test, y_train, y_test = train_test_split(X, data_y, test_size=0.2, random_state=15)
X_train
```

```python
#  w = (AᵀA)⁻¹Aᵀy
# training the model
A = np.array(X_train.values)
y = scaler.fit_transform(y_train.values.reshape(-1,1))
print("Training features shape: ",A.shape)
print("Training lables shape: ",y.shape)
W = np.linalg.inv(A.T @ A) @ (A.T @ y)
y_pred = A @ W
mse = mean_squared_error(y_pred=y_pred,y_true=y)
print("Mean square error for training: ",mse)
```

```python
# testing the model
A_test = X_test.values
y_validation = scaler.fit_transform(y_test.values.reshape(-1,1))
y_pred = A_test @ W
mse = mean_squared_error(y_pred=y_pred,y_true=y_validation)
print("Mean square error for testing: ",mse)
```
